```python
import pandas as pd
import os

# Definimos la ruta relativa
# .. (sale de notebooks) / data / raw / archivo.csv
ruta_datos = "../data/raw/partidos_cebollitas.csv"

# Verificamos si el archivo existe antes de cargarlo
if os.path.exists(ruta_datos):
    df = pd.read_csv(ruta_datos)
    print("✅ ¡Dataset cargado con éxito!")
    # Mostramos las primeras 5 filas
    display(df.head())
else:
    print(f"❌ No encontré el archivo en: {os.path.abspath(ruta_datos)}")
```

```python
df.describe()
```

```python
pgpplc = df[df['equipo_local']=='Cebollitas FC']['goles_local'].mean()
print(f"Promedio de gol por partido de local de Cebollitas FC: {pgpplc}")
```

```python
pgppvc = df[df['equipo_visitante']=='Cebollitas FC']['goles_visitante'].mean()
print(f"Promedio de gol por partido de visitante de Cebollitas FC: {pgppvc}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,4))

sns.histplot(df[df['equipo_local']=='Cebollitas FC']['goles_local'], kde=True, ax=ax[0], bins=10)
ax[0].set_title("Goles de Cebollitas de local")

sns.histplot(df[df['equipo_visitante']=='Cebollitas FC']['goles_visitante'], kde=True, ax=ax[1], bins=10)
ax[1].set_title("Goles de Cebollitas de visitante")
```

```python
sns.boxplot(x=df[df['equipo_local']=='Cebollitas FC']['goles_local'])
plt.title("Goles de local de Cebollitas FC")
```

```python
sns.scatterplot(x=df[df['equipo_local']=='Cebollitas FC']['posesion_local (%)'], y=df[df['equipo_local']=='Cebollitas FC']['goles_local'])
plt.title('posesion vs goles de local Cebollitas FC')
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[df['equipo_local']=='Cebollitas FC'][['goles_local','posesion_local (%)','tiros_arco_local']].corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de calor Cebollitas FC de local')
```
