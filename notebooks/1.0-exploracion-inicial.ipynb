```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuramos para que los gráficos se vean bonitos
sns.set_theme(style="whitegrid")
print("✅ Librerías cargadas correctamente")

# Creamos unos datos de prueba rápidos
df = pd.DataFrame({
    'x': np.random.randn(100),
    'y': np.random.randn(100)
})

# Graficamos algo rápido
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='x', y='y', color='teal')
plt.title("Mi primer gráfico en el entorno profesional")
plt.show()
```
