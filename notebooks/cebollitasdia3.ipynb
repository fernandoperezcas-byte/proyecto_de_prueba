```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler,StandardScaler
import os
```

```python
ruta_datos = "../data/raw/partidos_cebollitas.csv"

# Verificamos si el archivo existe antes de cargarlo
if os.path.exists(ruta_datos):
    df = pd.read_csv(ruta_datos)
    print("✅ ¡Dataset cargado con éxito!")
    # Mostramos las primeras 5 filas
    display(df.head())
else:
    print(f"❌ No encontré el archivo en: {os.path.abspath(ruta_datos)}")
```

```python
df['tiros_arco_local'].value_counts()
```

```python
plt.hist(df['tiros_arco_local'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Tiros al arco de local')
plt.ylabel('Frecuencia')
plt.show()
```

```python
scaler = MinMaxScaler()

df['tiros_arco_local_escalado'] = scaler.fit_transform(df[['tiros_arco_local']])

plt.hist(df['tiros_arco_local_escalado'], bins=10, color='blue', edgecolor='black')
```
